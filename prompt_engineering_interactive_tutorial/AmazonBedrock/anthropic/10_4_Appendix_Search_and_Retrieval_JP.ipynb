# Appendix 10.3: Search & Retrieval

（付録 10.3：検索とリトリーバル）

Claude を使って **Wikipedia を検索** できることをご存知ですか？ Claude は記事を見つけて取得でき、その後それらを要約・統合したり、得られた情報から新しい文章を書いたり、他にもさまざまなことができます。しかも Wikipedia だけではありません。プレーンテキストとして保存された自分のドキュメントでも、ベクターストアに埋め込んだドキュメントでも、検索対象にできます。

ベクターデータベース、Wikipedia、インターネットなどから取得したデータで Claude の知識を補い、応答の正確さや関連性を高める方法は、[RAG cookbook examples](https://github.com/anthropics/anthropic-cookbook/blob/main/third_party/Wikipedia/wikipedia-search-cookbook.ipynb) を参照してください。そこでは、特定の [embeddings](https://docs.anthropic.com/claude/docs/embeddings) やベクターデータベースツールの使い方も学べます。

Claude を使った高度な RAG アーキテクチャに興味がある場合は、[Claude 3 technical presentation slides on RAG architectures](https://docs.google.com/presentation/d/1zxkSI7lLUBrZycA-_znwqu8DDyVhHLkQGScvzaZrUns/edit#slide=id.g2c736259dac_63_782) も参照してください。
