### Anthropic SDK ã¨ Messages API
ã“ã®ãƒãƒ¥ãƒ¼ãƒˆãƒªã‚¢ãƒ«ã§ã¯ã€[Anthropic Python SDK](https://docs.anthropic.com/claude/reference/claude-on-amazon-bedrock) ã¨ [Messages API](https://docs.anthropic.com/claude/reference/messages_post) ã‚’ä½¿ç”¨ã—ã¾ã™ã€‚

ä»¥ä¸‹ã¯ã€ã“ã®ãƒãƒ¥ãƒ¼ãƒˆãƒªã‚¢ãƒ«ã§ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã‚’å®Ÿè¡Œã™ã‚‹ä¾‹ã§ã™ã€‚

ã¾ãšã€ãƒ¢ãƒ‡ãƒ«åã¨ãƒªãƒ¼ã‚¸ãƒ§ãƒ³ã‚’è¨­å®šã—ã¦ä¿å­˜ã—ã¾ã™ã€‚

```python
import boto3
session = boto3.Session() # create a boto3 session to dynamically get and set the region name
AWS_REGION = session.region_name
print("AWS Region:", AWS_REGION)
MODEL_NAME = "us.anthropic.claude-sonnet-4-20250514-v1:0"
# MODEL_NAME = "anthropic.claude-3-haiku-20240307-v1:0"

# https://docs.aws.amazon.com/ja_jp/bedrock/latest/userguide/models-supported.html

%store MODEL_NAME
%store AWS_REGION
```

æ¬¡ã«ã€`get_completion` ã‚’ä½œæˆã—ã¾ã™ã€‚ã“ã‚Œã¯ Claude ã«ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã‚’é€ä¿¡ã—ã€Claude ãŒç”Ÿæˆã—ãŸå¿œç­”ã‚’è¿”ã™ãƒ˜ãƒ«ãƒ‘ãƒ¼é–¢æ•°ã§ã™ã€‚ã“ã®ã‚»ãƒ«ã‚’å®Ÿè¡Œã—ã¦ãã ã•ã„ã€‚

```python
from anthropic import AnthropicBedrock

client = AnthropicBedrock(aws_region=AWS_REGION)

def get_completion(prompt, system=''):
    message = client.messages.create(
        model=MODEL_NAME,
        max_tokens=2000,
        temperature=0.0,
        messages=[
          {"role": "user", "content": prompt}
        ],
        system=system
    )
    return message.content[0].text
```

ãã‚Œã§ã¯ã€Claude å‘ã‘ã®ã‚µãƒ³ãƒ—ãƒ«ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆã‚’æ›¸ãã€`get_completion` ãƒ˜ãƒ«ãƒ‘ãƒ¼é–¢æ•°ã‚’ä½¿ã£ã¦ Claude ã®å‡ºåŠ›ã‚’è¡¨ç¤ºã—ã¦ã¿ã¾ã—ã‚‡ã†ã€‚æ¬¡ã®ã‚»ãƒ«ã‚’å®Ÿè¡Œã™ã‚‹ã¨ã€ã“ã®ã‚»ãƒ«ã®ä¸‹ã« Claude ã®å¿œç­”ãŒè¡¨ç¤ºã•ã‚Œã¾ã™ã€‚

ãƒ—ãƒ­ãƒ³ãƒ—ãƒˆæ–‡å­—åˆ—ã¯è‡ªç”±ã«å¤‰æ›´ã—ã¦ã€Claude ã®å¿œç­”ãŒã©ã†å¤‰ã‚ã‚‹ã‹è©¦ã—ã¦ã¿ã¦ãã ã•ã„ã€‚

```python
# Prompt
prompt = "Hello, Claude!"

# Get Claude's response
print(get_completion(prompt))
```
